```python
# Import the demand dataset and set the file path to it
# Define file paths for input and output
file_path = "OutputData/TR_2017_2023_MDT_2025-01-23.csv"
output_path = "OutputData/Aggregated_TR_2017_2023.csv"
```

The aggregate_by_basin function performs the following tasks:
1) Consolidates the mainstem (_M) and tributary basins; so R_02_M and R_02 records should be consolidated into R_02.
2) Aggregates the water right demand dataset at a basin level

Arguments:
    file_path (str): path to the input CSV file.
    output_path(str, optional): path to save the aggregated CSV file.

Returns:
    pd.DataFram: Aggregated DataFrame

```python
import pandas as pd

def aggregate_by_basin(file_path, output_path=None):

df = pd.read_csv(file_path)

    # Define the columns to aggregate
    columns_to_aggregate = [
        "JAN_MEAN_DIV", "FEB_MEAN_DIV", "MAR_MEAN_DIV", "APR_MEAN_DIV",
        "MAY_MEAN_DIV", "JUN_MEAN_DIV", "JUL_MEAN_DIV", "AUG_MEAN_DIV",
        "SEP_MEAN_DIV", "OCT_MEAN_DIV", "NOV_MEAN_DIV", "DEC_MEAN_DIV"
    ]

    # Remove the '_M' suffix from the BASIN column
    df["BASIN"] = df["BASIN"].str.replace("_M$", "", regex=True)
    
    # Group by the cleaned BASIN column and sum the selected columns
    aggregated_df = df.groupby("BASIN")[columns_to_aggregate].sum().reset_index()
    
    # Save the result to a file if output_path is provided
    if output_path:
        aggregated_df.to_csv(output_path, index=False)
    
    return aggregated_df
```
